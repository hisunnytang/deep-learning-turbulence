# generating power spectrum for a given spectral index

\begin{equation}
P_k = < V^{*}(k) V(k)> \approx E_0 k^{-\beta} \\
V(k) = \int V(x) e^{ikx} d^3 x \\
\approx \sqrt{E_0 k^{-\beta}} e^{i \theta}  \\
\end{equation}

where $\theta$ is the random phase factor. Inverse fourier transform of the above $V(k)$ gives the $V(x)$.

However, in real life, we don't usually have info on the line-of-sight dimension.
What we see is instead, $I(v)$ intensity as a function of velocity bins, i.e. historgram of the V(x) at different velocity bins. Our goal here is to see to what extent can we extract that the missing information in turbulence from the intensity spectrum $I(v)$.

```python
def power_spectrum( E0, beta, kmin = 1.0, Nx = 1024):
    return lambda k: ( (k / kmin)**-beta ) * ( - 1.0 + beta)/ kmin


def get_real_space_signal( beta,  Nx = 1024,  Nbins = 128 ):
    x    = np.linspace(0, 1, Nx)
    bins = np.linspace( -1.0, 1.0, Nbins + 1)
    kx =  x * Nx * 2.0*np.pi
    kmin = kx[1]
    Vx = np.zeros((len(x)))
    
    Pk = power_spectrum(1.0 , beta, kmin=kmin)
    
    for k_ in kx[1:]:
        Pk_ = Pk( k_ )
        Vk_ = np.sqrt( Pk_)
        phi = 2.0 * np.pi * random(1)
        ans = Vk_ * np.cos( k_ * x + phi  )   
#         ans -= numpy.mean(ans)
#         plt.plot( x, ans )
        Vx += ans.real
#     plt.show()
    Vx -= np.mean(Vx)
    count, _ = np.histogram(Vx, bins=bins)
#     print( "range of Vx: min = {}, max = {} with beta = {}".format(min(Vx), max(Vx), beta )  )
    return Vx, count
```

```python
import numpy as np
Vx_b = []
count_b = []
Nbeta = 4
bins = np.linspace( -1.0, 1.0, 128 + 1)
x    = np.linspace(0, 1, 1024)
beta_b = numpy.linspace( 1.1, 4, Nbeta )
for b in beta_b:
    Vx, count = get_real_space_signal( b )
    Vx_b.append(Vx)
    count_b.append(count)

for i in range(Nbeta):
    plt.plot( (bins[1:] + bins[:-1] ) /2.0, count_b[i] / max(count_b[i]), label ="{}".format(beta_b[i]) )
plt.xlabel('velocity')
plt.ylabel('Intensity')
plt.legend()
plt.show()

for i in range(Nbeta):
    plt.plot(x, Vx_b[i] / max(Vx_b[i]), label = "{}".format(beta_b[i]) )
plt.ylabel('Velocity')
plt.xlabel('x')
plt.legend()
plt.show()

for i in range(Nbeta):
    plt.loglog( range(127), numpy.abs( numpy.fft.fft( Vx_b[i] )**2.0 )[1:128], label ="{}".format(beta_b[i]) )
plt.ylabel('Power Spectrum from our synthetic $V(x)$')
plt.xlabel('k = 1/ x (wavenumber)')
plt.legend()
plt.show()
#     plt.loglog( range(500), 1.0e5 * numpy.arange(500)**-beta_b[i] )
```

```python
def get_batch( X, y, batch_size = 128 ):
    
    a   = numpy.arange(len(y))
    numpy.random.shuffle(a)

    batch_x = X[a[:batch_size]]
    batch_y = y[a[:batch_size]]

    batch_x = numpy.array(batch_x, dtype=numpy.float32)
    batch_y = numpy.array(batch_y, dtype=numpy.float32)

    for i in range(batch_size):
        batch_x[i] = batch_x[i]/ batch_x[i].max()
    
    
    return batch_x[:,:,None], batch_y
```
